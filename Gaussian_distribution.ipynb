```python

import numpy as np
import matplotlib.pyplot as plt

n=1000
u1=np.random.rand(n)
u2=np.random.rand(n)
z0 = np.sqrt(-2 * np.log(u1)) * np.cos(2 * np.pi * u2)
z1 = np.sqrt(-2 * np.log(u1)) * np.sin(2 * np.pi * u2)
mean = 0
std_dev = 1


gaussian_data = mean + std_dev * z0
plt.hist(gaussian_data, bins=50, density=True, alpha=0.6, color='g')


x = np.linspace(min(gaussian_data), max(gaussian_data), 1000)
plt.plot(x, (1/(np.sqrt(2*np.pi)*std_dev)) * np.exp(-0.5*((x-mean)/std_dev)**2), linewidth=2, color='r')



plt.title("Gaussian Distribution (Generated using Box-Muller Transform)")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.show()
```
